```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns   
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('df_preprocessed.csv')
df
```

```python
X = df.drop('max_monthly_emi', axis=1)
X.shape
```

```python
y = df['max_monthly_emi']
y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
```

```python
RandomForestRegressor()
```

```python
RF = RandomForestRegressor(
    n_estimators=200,         # number of trees
    max_depth=15,             # limit depth to prevent overfitting
    min_samples_split=5,      # min samples to split an internal node
    min_samples_leaf=3,       # min samples at leaf node
    max_features='sqrt',      # number of features to consider at each split
    bootstrap=True,           # use bootstrap samples
    n_jobs=-1 )
```

```python
RF.fit(x_train, y_train)
```

```python
RF.score(x_test, y_test)
```

```python
df_non_tree = pd.read_csv('df_transformed.csv')
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline

X_1 = df_non_tree.drop('max_monthly_emi', axis=1)
y_1 = df_non_tree['max_monthly_emi']

```

```python
x_train_1, x_test_1, y_train_1, y_test_1 = train_test_split(X_1, y_1, test_size=0.2, random_state=21)
```

```python
SVR()
```

```python
svm_ = Pipeline([
    ('scaler', MinMaxScaler()),
    ('svm', SVR(kernel='rbf', C=10, gamma='scale', epsilon=0.05, max_iter=2000))
])

svm_.fit(x_train_1, y_train_1)

svm_.score(x_test_1, y_test_1)
```

```python
svm_unscaled =SVR(kernel='rbf', C=10, gamma='scale', epsilon=0.05, max_iter=2000)


svm_unscaled.fit(x_train_1, y_train_1)

svm_unscaled.score(x_test_1, y_test_1)
```

```python
LinearRegression()
```

```python
linear_ = Pipeline([
    ('scaler', MinMaxScaler()),
    ('LogisticRegression', LinearRegression(n_jobs=-1 ))
])

linear_.fit(x_train_1, y_train_1)

linear_.score(x_test_1, y_test_1)
```

```python
linear_unscaled = Pipeline([
    ('scaler', MinMaxScaler()),
    ('LogisticRegression', LinearRegression(n_jobs=-1 ))
])

linear_unscaled.fit(x_train_1, y_train_1)

linear_unscaled.score(x_test_1, y_test_1)
```

```python
linear_untransformed = Pipeline([
    ('scaler', MinMaxScaler()),
    ('LogisticRegression', LinearRegression(n_jobs=-1 ))
])

linear_untransformed.fit(x_train, y_train)

linear_untransformed.score(x_test, y_test)
```

```python
XGBRegressor()
```

```python
xgb_ = XGBRegressor(
    n_estimators=300,       
    learning_rate=0.05,     
    max_depth=6,            
    subsample=0.8,          
    colsample_bytree=0.8,   
    reg_alpha=0.1,           
    reg_lambda=1,            
    random_state=42,
    n_jobs=-1
)
```

```python
xgb_.fit(x_train, y_train)
```

```python
xgb_.score(x_test, y_test)
```

```python
y_pred_xgb = xgb_.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, root_mean_squared_error

print("R2:", r2_score(y_test, y_pred_xgb))
print("MAE:", mean_absolute_error(y_test, y_pred_xgb))
print("RMSE:", root_mean_squared_error(y_test, y_pred_xgb))
```
